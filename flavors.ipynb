```python
import pandas as pd

file_path = '/Users/kseniaklimova/Documents/GitHub/wine_website/wine-gallery/public/data/updated_wine_data.csv'
df = pd.read_csv(file_path)

# Display the dataframe to check
df[['description', 'extracted_flavors']].head()
```

```python
# Final flavor_categories dictionary
flavor_categories = {
    "wood": ["oak", "cedarwood", "burnt", "roasted", "wood", "charcoal", "sandal", "sandalwood"],
    "smoky": ["smoke", "tobacco", "gunsmoke", "cured", "baked", "grilled", "charcoal", "bacon", "smoked"],
    "vanilla": ["vanilla", "buttercream"],
    "butter": ["butter", "buttercream", "custard"],
    "citrus": ["lemon", "lime", "orange", "citrus", "grapefruit", "tangerine", "zest", "kiwi"],
    "berry": ["berry", "cherry", "currant", "red currant", "raspberry", "blueberry", "strawberry", "blackberry", "mulberry", "cranberry", "jam", "cassis"],
    "leather": ["leather"],
    "nuts": ["nut", "almond", "hazelnut"],
    "coffee": ["coffee", "cocoa", "chocolate", "espresso"],
    "herbs": ["herb", "wintergreen", "lemongrass", "sage", "grassy", "grass", "balsamic", "menthol", "thyme"],
    "spice": ["spice", "pepper", "licorice", "ginger", "cinnamon", "clove", "anise", "saffron"],
    "fruit": ["apple", "pear", "peach", "melon", "tropical", "plum", "prune", "grapes", "grape", "pineapple", "figs", "apricot", "quince"],
    "chocolate": ["chocolate", "mocha", "cocoa", "milk chocolate"],
    "mineral": ["mineral", "earth", "minerality", "minerals", "brick", "stone"],
    "caramel": ["maple", "honey", "custard", "candy", "sugar", "jelly", "candied"],
    "flowers": ["lavender", "flower", "floral", "blossom", "rose", "petal"]
}

# Function to extract general flavor categories
def extract_general_flavors(description):
    description_lower = description.lower()  # Convert the description to lowercase
    matched_categories = set()  # Use a set to avoid duplicates
    for category, terms in flavor_categories.items():
        if any(term in description_lower for term in terms):
            matched_categories.add(category)
    return list(matched_categories)  # Return as a list

# Apply the function to the "description" column
df['extracted_flavors'] = df['description'].apply(extract_general_flavors)

# Display a sample of the data with the extracted general flavors
df[['description', 'extracted_flavors']].head(10)
```
