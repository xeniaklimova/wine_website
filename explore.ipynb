```python

wine_type_keywords = {
    "red": ["cabernet", "merlot", "pinot noir", "syrah", "zinfandel", "malbec", "red", "barbera",
            "aglianico", "gamay", "grenache", "nebbiolo", "tempranillo", "shiraz", "sangiovese"],
    "white": ["chardonnay", "sauvignon blanc", "riesling", "pinot grigio", "white", "viognier",
              "gewürztraminer", "grüner veltliner", "pinot gris"],
    "sparkling": ["sparkling", "champagne", "prosecco", "cava"],
    "rosé": ["rosé", "rose"],
    "dessert": ["dessert", "port", "sherry", "sauternes", "moscato"],
    "fortified": ["fortified", "port", "sherry", "madeira"]
}

def determine_wine_type(title, variety):
    # Convert title and variety to lowercase for case-insensitive matching
    title_lower = title.lower() if isinstance(title, str) else ""
    variety_lower = variety.lower() if isinstance(variety, str) else ""

    # Check for keywords in both title and variety
    for wine_type, keywords in wine_type_keywords.items():
        if any(keyword in title_lower for keyword in keywords) or \
           any(keyword in variety_lower for keyword in keywords):
            return wine_type
    # Default to "unknown" if no type is matched
    return "unknown"

df['wine_type'] = df.apply(
    lambda row: determine_wine_type(row['title'], row['variety']), axis=1
)

# Display a sample of the data with the wine type
df[['title', 'variety', 'wine_type']].head(10)
```

```python
print(df[df['wine_type'] == 'unknown'].shape[0])
```

```python
df.dropna(subset=['price'], inplace=True)
df.dropna(subset=['year'], inplace=True)
df = df[df['wine_type'] != 'unknown']
df['year'] = df['year'].astype(int)

df.head(30)
```

```python
print(df.shape[0])

flavor_terms = [
    "fruit", "berry", "citrus", "apple", "pear", "peach", "cherry",
    "tropical", "lemon", "lime", "chocolate", "vanilla", "oak",
    "spice", "pepper", "nut", "herb", "floral", "honey",
    "butter", "smoke", "earth", "mineral", "almond", "leather", "coffee", "lemon", "orange",
    "cocoa", "plum", "prune"
]

# add another column: soft, rich, dry, tannic, sweet, bold, light, acidic, tart, heavy, smooth

```

```python
def extract_flavors(description):
    # Convert the description to lowercase
    description_lower = description.lower()
    # Find and return flavor terms that appear in the description
    matched_terms = [term for term in flavor_terms if term in description_lower]
    return matched_terms

# Apply the function to the "description" column
df['extracted_flavors'] = df['description'].apply(extract_flavors)

# Display a sample of the data with the extracted flavors
df[['description', 'extracted_flavors']].head(10)
                            
```

```python
df['price'] = df['price'].astype(int)
df.drop(columns=['Unnamed: 0'])
df.to_csv('final.csv', index=True)
new_df = pd.read_csv('final.csv')
```
